This is just a notebook to invoke AWS Sagemaker training jobs remotely. Means I'm only running hte ml.p3.2xlarge while a job is training and not in a notebook for an extended period.

```python
import sys
import logging

nblog = open("nb.log", "a+")
sys.stdout.echo = nblog
sys.stderr.echo = nblog

get_ipython().log.handlers[0].stream = nblog
get_ipython().log.setLevel(logging.INFO)

%autosave 5
```

```python
import sagemaker
from sagemaker.pytorch import PyTorch
from sagemaker import get_execution_role

sagemaker_session = sagemaker.Session()

output_path = f"s3://phishing-edge/phishing-edge-output"  # Where to save model artifacts
role = get_execution_role()

git_config = {'repo': 'https://github.com/imack/phishing-edge.git', 'branch': 'main'}


estimator = PyTorch(
    entry_point='train.py',
    role=role,
    source_dir='./',
    instance_count=1,
    instance_type='ml.p3.2xlarge',
    framework_version='2.1.0',
    py_version='py310',
    git_config=git_config,
    output_path=output_path,
    hyperparameters={
        'epochs': 50,
        'batch_size': 32,
        'lr': 0.001
    },
    volume_size=100,
)

estimator.fit()
```
